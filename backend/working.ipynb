```python
import ast

def convert(obj):
    L = []

    obj = ast.literal_eval(obj)

    for item in obj:
        L.append(item["name"])

    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies["genres"].head()
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies["cast"].iloc[0]
```

```python
def convert3(obj):
    L = []
    obj = ast.literal_eval(obj)
    count = 0

    for item in obj:
        L.append(item["name"])
        count += 1

        if count == 3:
            break

    return L
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies[["title", "cast"]].head()
```

```python
movies['crew'].head()
```

```python
def fetch_director(obj):
    L = []

    obj = ast.literal_eval(obj)

    for item in obj:
        if(item['job']== "Director"):
            L.append(item["name"])
        
    return L;
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies[["title", "genres", "keywords", "cast", "crew"]].head()
```

```python
movies[movies['overview'].isnull()]
```

```python
movies['overview'] = movies['overview'].apply(
    lambda x: x.split() if isinstance(x, str) else []
)
```

```python
movies['genres'] = movies['genres'].apply(
    lambda x: [i.replace(" ", "") for i in x]
)
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew']
)
```

```python
new_df = movies[['movie_id', 'title', 'tags']]  
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'].iloc[0]
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity.shape
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]

    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
# V2 -of project - multiple movie input 

liked_movies = [
    'Iron Man 2',
    'The Dark Knight',
    'The Avengers'
    'Man of Steel'
]
```

```python
movie_indices = []
missing_movies = []

for movie in liked_movies:
    matches = new_df.index[
        new_df["title"].str.strip().str.casefold() == movie.strip().casefold()
    ]

    if not matches.empty:
        movie_indices.append(matches[0])
    else:
        missing_movies.append(movie)

print("Missing movies:", missing_movies)
```

```python
liked_vectors = []

for idx in movie_indices:
    liked_vectors.append(vectors[idx])
```

```python
import numpy as np;
```

```python
user_vector = np.mean(liked_vectors, axis=0)
```

```python
user_vector = user_vector.reshape(1, -1)
```

```python
distances = cosine_similarity(
    user_vector,
    vectors
)
```

```python
movies_list = list(enumerate(distances[0]))
movies_list = sorted(
    movies_list,
    reverse=True,
    key=lambda x: x[1]
)
```

```python
# Remove already selected movies
for i in movies_list:
    movie_idx = i[0]

    if movie_idx not in movie_indices:
        print(new_df.iloc[movie_idx].title)
```
